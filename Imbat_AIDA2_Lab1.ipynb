```python
grades_dict = pd.read_csv('/content/grades.csv', header=0, squeeze=True).to_dict()
classlist1 = pd.read_csv('/content/classlist_1.csv').to_dict()
classlist2 = pd.read_csv('/content/classlist_2.csv').to_dict()

class Section:
    def __init__(self, grades, classlist_1, classlist_2):
     self.grades = grades
     self.classlist_1 = classlist_1
     self.classlist_2 = classlist_2
  
    def get_grades(self):
      self.gradesDF = pd.DataFrame(self.grades)
      self.gradesDF['Sem Grade'] = (self.gradesDF['prelim_g']*.30) + (self.gradesDF['mid_g']*.30) + (self.gradesDF['final_g']*.40)
      self.gradesDF['name'] = self.gradesDF['name'].str[0] + self.gradesDF['name'].str[1:].str.replace('[a-z]', '*')
      return (self.gradesDF)
    
    def get_class_stats(self):
      prelim = np.array(list(self.gradesDF['prelim_g'].values()))
      midterm = np.array(list(self.gradesDF['midterm_g'].values()))
      finals = np.array(list(self.gradesDF['final_g'].values()))
      semestral = np.array(list(self.gradesDF['Sem Grade'].values()))

      min_max = {'Lowest': {'Prelim': prelim.min(), 'Midterm':midterm.min(), 
                            'Finals':finals.min(), 'Semestral':semestral.min()},
                'Highest': {'Prelim': prelim.max(), 'Midterm':midterm.max(), 
                            'Finals':finals.max(), 'Semestral':semestral.max()},
                'Mean': {'Prelim': np.mean(prelim), 'Midterm':np.mean(midterm), 
                          'Finals':np.mean(finals), 'Semestral':np.mean(semestral)},
                'Median': {'Prelim': np.median(prelim), 'Midterm':np.median(midterm), 
                            'Finals':np.median(finals), 'Semestral':np.median(semestral)},
                'Mode': {'Prelim': st.mode(prelim), 'Midterm':st.mode(midterm), 
                          'Finals':st.mode(finals), 'Semestral':st.mode(semestral)},
                'STD': {'Prelim': np.std(prelim), 'Midterm':np.std(midterm), 
                        'Finals':np.std(finals), 'Semestral':np.std(semestral)}}
      return (self.min_max)

    def get_similar_students(self):
      set1 = set(l for l in self.classlist_1)
      return [l for l in self.classlist_2 if l in set1]

      print(get_similar_students(self.classlist_1, self.classlist_2))

    def get_failed(self):
      self.gradesDF['Remark'] = np.where(self.gradesDF['Sem Grade'] < 70, 'Failed', 'Passed')
      self.failed_df = self.gradesDF[(self.gradesDF['Remark']=='Failed')]
      return self.failed_df
    
    def fail_count(self):
      count = len(self.failed_df)
      return count
```

```python
section = Section(grades_dict, classlist1, classlist2)
grades = section.get_grades()
similar = section.get_similar_students()
failed = section.get_failed()
failedNum = section.fail_count()
```
